```python
import os

def list_files_and_sizes(start_path):
    for root, dirs, files in os.walk(start_path):
        print(f'Directory: {root}')
        total_size = 0
        for file in files:
            file_path = os.path.join(root, file)
            # Ignore if it's not a file.
            if not os.path.isfile(file_path):
                continue
            file_size = os.path.getsize(file_path)
            total_size += file_size
            print(f'\t{file} Size: {file_size} bytes')
        print(f'Total size of files in directory: {total_size} bytes\n')
```

```python
list_files_and_sizes('./results/raw_results/fed_imp_pc2/alpha/')
```
